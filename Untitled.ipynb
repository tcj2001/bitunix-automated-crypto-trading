```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

def calculate_best_fit_line(y_values, x_values=None):
    """Calculates the best fit line (linear regression) for the given y values."""
    if x_values is None:
        x_values = np.arange(len(y_values))
    slope, intercept = np.polyfit(x_values, y_values, 1)
    return slope, intercept
```

```python
def plot_trend_lines(ticker="AAPL", lookback=20):
    """
    Fetches historical data and plots trend lines tilted from the highest high
    and lowest low to align with subsequent price action.
    """
    data = yf.download(ticker, period="240d")  # Adjust period as needed
    if data.empty or len(data) < lookback:
        print(f"Not enough data for {ticker} with a lookback of {lookback} candles.")
        return

    recent_data = data[-lookback:]
    high_prices = recent_data['High'].values
    low_prices = recent_data['Low'].values
    midpoints = (high_prices + low_prices) / 2
    candle_indices = np.arange(lookback)

    # 1. Identify highest high and lowest low candle
    highest_high_index = np.argmax(high_prices)
    lowest_low_index = np.argmin(low_prices)
    highest_high_price = high_prices[highest_high_index]
    lowest_low_price = low_prices[lowest_low_index]

    # 2. Top line: Tilt down from highest high to touch a subsequent high (if any)
    slope_top = -float('inf')  # Initialize with a very steep negative slope
    intercept_top = highest_high_price - slope_top * highest_high_index

    for i in range(highest_high_index + 1, lookback):
        current_slope = (high_prices[i] - highest_high_price) / (i - highest_high_index) if (i - highest_high_index) != 0 else 0
        if current_slope > slope_top:
            slope_top = current_slope
            intercept_top = highest_high_price - slope_top * highest_high_index

    top_line = slope_top * candle_indices + intercept_top

    # 3. Bottom line: Tilt up from lowest low to touch a subsequent low (if any)
    slope_bottom = float('inf')  # Initialize with a very steep positive slope
    intercept_bottom = lowest_low_price - slope_bottom * lowest_low_index

    for i in range(lowest_low_index + 1, lookback):
        current_slope = (low_prices[i] - lowest_low_price) / (i - lowest_low_index) if (i - lowest_low_index) != 0 else 0
        if current_slope < slope_bottom:
            slope_bottom = current_slope
            intercept_bottom = lowest_low_price - slope_bottom * lowest_low_index

    bottom_line = slope_bottom * candle_indices + intercept_bottom

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(candle_indices, high_prices, label='High', color='green')
    plt.plot(candle_indices, low_prices, label='Low', color='red')
    plt.plot(candle_indices, midpoints, label='Midpoint', color='blue', linestyle='--')
    plt.scatter(highest_high_index, highest_high_price, color='green', marker='o', label='Highest High')
    plt.scatter(lowest_low_index, lowest_low_price, color='red', marker='o', label='Lowest Low')
    plt.plot(candle_indices, top_line, label='Top Line (Tilted from High)', color='darkgreen', linewidth=2)
    plt.plot(candle_indices, bottom_line, label='Bottom Line (Tilted from Low)', color='darkred', linewidth=2)

    plt.title(f'{ticker} Last {lookback} Candles with Tilted Trend Lines')
    plt.xlabel('Candle Index (Recent)')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.xticks(candle_indices)
    plt.tight_layout()
    plt.show()

    print("\nInterpretation:")
    print("The top line starts at the highest high and tilts downwards to connect with the lowest slope encountered at subsequent highs.")
    print("The bottom line starts at the lowest low and tilts upwards to connect with the highest slope encountered at subsequent lows.")
    print("This approach attempts to create trend lines that are anchored at the extremes and aligned with the flow of subsequent price action.")

# Example usage:
#plot_trend_lines(ticker="AAPL", lookback=20)
# You can adjust the ticker and lookback period as needed
plot_trend_lines(ticker="GOOGL", lookback=10)
```
